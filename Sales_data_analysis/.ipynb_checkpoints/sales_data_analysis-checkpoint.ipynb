```python
import pandas as pd
import os 
```

### Merging monthly data into single file

```python
files = [file for file in os.listdir('./Pandas-Data-Science-Tasks-master/SalesAnalysis/Sales_Data')]

all_data = pd.DataFrame()
for file in files:
    df = pd.read_csv("./Pandas-Data-Science-Tasks-master/SalesAnalysis/Sales_Data/"+file)
    all_data = pd.concat([all_data, df])

all_data.to_csv('Cleaned_data/all_sales_data', index = False)
```

### Read from updated file

```python
data = pd.read_csv('Cleaned_data/all_sales_data')
data.head()
```

### Clean up data

```python
missing_data = data.isnull()
missing_data
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
nan_df = data[data.isna().any(axis=1)]
nan_df.shape
```

```python
data = data.dropna(how='all')
data
```

```python
temp_df = data[data['Order Date'].str[0:2] == 'Or']
temp_df
```

### Augment data with additional columns

```python
data = data[data['Order Date'].str[0:2] != 'Or']
data
```

### Task 2 : Add month column

```python
data['Month'] = data['Order Date'].str[0:2]
data['Month'].astype('int32')
data
```

### Task 3: Add a sales column

```python
data.dtypes
```

```python
data[["Order ID","Quantity Ordered"]] = data[["Order ID","Quantity Ordered"]].astype('int32')
data[["Price Each"]] = data[["Price Each"]].astype('float32')
```

```python
data.dtypes
```

```python
data["Sales"] = data['Quantity Ordered'] * data['Price Each']
data
```

### Question 1: What was the best month for sales? How much was earned that month?

```python
monthly_sales = data[['Month','Sales']].groupby('Month').sum()
monthly_sales
```

```python
import matplotlib.pyplot as plt

months = range(1,13)
plt.bar(months, monthly_sales['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD($)')
plt.xlabel('Month number')
plt.show()
```

### Task 4: Add a city column

```python
data['City'] = data['Purchase Address'].apply(lambda x: x.split(',')[1])
data.head()
```

```python
data
```

### Question 2: What city had the highest number of sales?

```python
sales_by_city = data[['City','Sales']].groupby('City').sum()
sales_by_city
```

```python
cities = [city for city, df in data.groupby('City')]

plt.bar(cities, sales_by_city['Sales'])
plt.xticks(cities, rotation = 'vertical')
plt.ylabel('Sales in USD($)')
plt.xlabel('City Name')
plt.show()
```

### Question 3: What time we should display advertisements to maximize likelihood of customer's buying product?

```python
data['Order Date'] = pd.to_datetime(data['Order Date']) 
```

```python
data.head()
```

```python
data['Hour'] = data['Order Date'].dt.hour
data['Minute'] = data['Order Date'].dt.minute
data
```

```python
hours = [hour for hour, df in data.groupby('Hour')]

plt.plot(hours, data.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel('Hour')
plt.ylabel('Number of orders')
plt.grid()
plt.show()
```

### Question 4: What products are most often sold together?

```python
df_sold_together = data[data['Order ID'].duplicated(keep=False)]

df_sold_together['Grouped'] = df_sold_together.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df_sold_together = df_sold_together[['Order ID', 'Grouped']].drop_duplicates()
df_sold_together.head(100)
```

```python
from itertools import combinations
from collections import Counter 
```

```python
count = Counter()

for row in df_sold_together['Grouped']:
    row_list = row.split()
    count.update(Counter(combinations(row_list, 2)))
    
count.most_common()
    
```

### Question 5: What product sold the most? Why do you think it sold the most?

```python
df_product_group = data.groupby('Product')
quantity_ordered = df_product_group.sum()['Quantity Ordered']

products = [product for product, df in df_product_group]

plt.bar(products, quantity_ordered)
plt.ylabel('Quantity Ordered')
plt.xlabel('Product')
plt.xticks(products, rotation = 'vertical', size = 8)
plt.show()
```

```python
prices = data.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered)
ax2.plot(products, prices, 'g-')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color = 'b')
ax2.set_ylabel('Price ($)', color = 'g')
ax1.set_xticklabels(products, rotation='vertical', size=8)

plt.show()
```
