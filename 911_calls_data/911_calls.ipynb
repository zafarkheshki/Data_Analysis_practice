# 911 Calls Data Analysis

This practice project is to analyze some 911 call data from [Kaggle] (https://www.kaggle.com/mchirico/montcoalert). The data contains the following fields:

* lat : Latitude
* lng: Longitude
* desc: Description of the Emergency Call
* zip: Zipcode
* title: Title
* timeStamp: YYYY-MM-DD HH:MM:SS
* twp: Township
* addr: Address
* e: Dummy variable (always 1)

## Creating new features

<b>Creating Departments columns to see stats of calls to each department.

```python
df['Department'] = df['title'].apply(lambda title: title.split(':')[0])
```

```python
df['Department'].value_counts()
```

```python
sns.countplot(x = 'Department', data=df)
```

<b>Creating Hour, Month, and Day of Week columns from timeStamp to explore more about emergency calls.

```python
df['Hour'] = df['timeStamp'].apply(lambda time: time.hour)
df['Month'] = df['timeStamp'].apply(lambda time: time.month)
df['Day of Week'] = df['timeStamp'].apply(lambda time: time.dayofweek)
```

```python
df
```

Need to map days of week from numbers to names.

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df['Day of Week'] = df['Day of Week'].map(dmap)
```

```python
df
```

Let's explore days of week firrst.

```python
sns.countplot(x='Day of Week', data = df, hue = 'Department')
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad=0.)
```

Doing same for months.

```python
sns.countplot(x='Month', data = df, hue = 'Department')
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad=0.)
```

In the plot, months 9,10, and 11 are missing. Let's explore our data if we have those months in our data or they do not exist.

```python
df[df['Month']==9|10|11]
```

We do not have data entries for the months 9, 10, and 11

```python
byMonth = df.groupby('Month').count()
byMonth.head()
```

```python
byMonth['title'].plot()
```

Creating a new Date column

```python
df['Date'] = df['timeStamp'].apply(lambda t: t.date())
```

```python
plt.figure(figsize=(12,6))
df.groupby('Date').count()['title'].plot()
```

Now, exploring every department individually.

```python
plt.figure(figsize=(12,6))
df[df['Department']=='Traffic'].groupby('Date').count()['title'].plot()
plt.title('Traffic')
```

```python
plt.figure(figsize=(12,6))
df[df['Department']=='Fire'].groupby('Date').count()['title'].plot()
plt.title('Fire')
```

```python
plt.figure(figsize=(12,6))
df[df['Department']=='EMS'].groupby('Date').count()['title'].plot()
plt.title('EMS')
```
